```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
examples = np.array([[1, 2], [2, 1], [3, 3], [0, 0], [-1, -1], [-3, -1]])
print(examples.shape)
```

```python
labels = np.array([1,1,1,-1,-1,-1])
```

```python
plt.scatter(examples[:, 0], examples[:, 1], c=labels, s=50)
```

```python
svc_example=SVC(kernel='linear',C=1000,shrinking=False)
svc_example.fit(examples, labels)
print('Indices of support vectors = ', svc_example.support_)
print('Support vectors =\n', svc_example.support_vectors_)
```

```python
def plot_svc_decision_function2(model, ax=None, plot_support=True):
    if ax is None:
        ax = plt.gca()


    x = np.linspace(-4, 4, 30)
    y = np.linspace(-2, 4, 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    ax.scatter(examples[:, 0], examples[:, 1], c=labels, s=50)
    
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
```

```python
plot_svc_decision_function2(svc_example,plot_support=False)
```
